```python
import pandas as pd
import matplotlib

data = pd.read_csv('mushrooms.csv', sep=',')

print(data[:3])
```

```python
print(data.describe())
```

```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline

for columnName in data.columns.values:
    uniques, classNames = np.unique(data['class'], return_inverse=True)
    uniques, column = np.unique(data[columnName], return_inverse=True)
    x = classNames
    y = column
    print(uniques)
    plt.ylabel(columnName)
    plt.scatter(x, y)
    plt.show()
```

```python
tmp = np.delete(np.asarray(data[data.columns]), [0], 1)
print(tmp)

X = []
for i in tmp:
    arr = []
    for j in i:
        arr.append(float(ord(j) - 96)/26)
    X.append(arr)
print(X)
```

```python
tmp = np.asarray(data[data.columns[0]])
print(tmp)
Y = []
for i in tmp:
    if i == 'p':
        Y.append(1.0)
    else:
        Y.append(0.0)
print(Y)
```

```python
# Create first network with Keras
from keras.models import Sequential
from keras.layers import Dense
import numpy
import pandas as pd
import matplotlib

# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
# load pima indians dataset
# dataset = numpy.loadtxt("iris.txt", delimiter=",")
#dataset = numpy.loadtxt("pima-indians-diabetes.data", delimiter=",")
# split into input (X) and output (Y) variables
# X = dataset[:,0:4]
# Y = dataset[:,4:7]
# create model
model = Sequential()
model.add(Dense(12, input_dim=22, init='uniform', activation='relu'))
model.add(Dense(8, init='uniform', activation='relu'))
model.add(Dense(1, init='uniform', activation='sigmoid'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Fit the model
model.fit(X, Y, epochs=10, batch_size=10,  verbose=2)
# calculate predictions
predictions = model.predict(X)
```

```python

rounded = [np.round(x) for x in predictions]
print(rounded)
```

```python
d = [[0, 0], [0, 0]]
def check(index, x):
    ans = (Y[index] == x)
    if x == 0.0:
        if ans:
            d[0][0] += 1;
        else:
            d[0][1] += 1;
    else:
        if ans:
            d[1][1] += 1;
        else:
            d[1][0] += 1;
    return 
result = [check(index, x) for index, x in enumerate(rounded)]

print(d)
```
